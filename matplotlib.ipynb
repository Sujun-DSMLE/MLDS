```python
# Importing
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
# %matplotlib notebook
```

```python
# Data
ages = np.linspace(0, 1, 100)
heights = np.sin(ages)
health = np.cos(heights)
```

```python
# Line Plot
plt.plot(ages, heights, label="AgesVsHeights", c="darkorange", marker='o')
plt.plot(ages, health, label="AgesVsHealth", c="lightgreen", marker='^')
plt.xlabel("Ages")
plt.ylabel("Heights and Health")
plt.legend()
plt.grid(True)
plt.xlim(0,1.1)
plt.ylim(0, 1.1)
plt.yticks([0.0, 0.5, 1.0])
plt.xticks([0.1, 0.25, 0.50, 0.75, 0.1])
plt.savefig("line.jpg")
plt.show()
```

```python
# Area Plot
plt.fill_between(ages, heights)
plt.fill_between(ages, health)
plt.show()
```

```python
plt.stackplot(ages, heights)
plt.stackplot(ages, heights-0.1)
plt.show()
```

```python
#Bar Plot
x = list(range(1,5))
y = [5,3,3,5]
z = [0.6, 2, 1.5, 4]
plt.bar(x, y, color="darkorange")
plt.barh(x,z, color="lightgreen")
plt.title("XvsY")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
# Histogram
bins = [1,2,3,4,5,6]
plt.hist(y, bins=bins)
plt.show()
```

```python
# Scatter Plot
plt.scatter(ages[::3],heights[::3], color="darkorange", marker="^")
plt.show()
```

```python
# Pie Chart
categories = ["Poetry", "Cricket","Volleyball","Athletics"]
participation = [23, 77, 55, 101]
plt.pie(participation, labels=categories, explode=(0.1,0,0,0), shadow=(0,0,0,0.5))
plt.show()
```

```python
# 3D Plot

fig= plt.figure()
axes = plt.axes(projection = "3d")
axes.plot3D(ages, heights, health)
axes.set_xlabel("Ages")
axes.set_ylabel("Heights")
axes.set_zlabel("Health")
plt.show()
```

```python
# Plot Live Data
X=[]
Y=[]
for i in range(100):
    X.append(ages[i])
    Y.append(np.sin(ages[i]))
    plt.scatter(X,Y, color="darkorange", marker="^")
    plt.pause(0.5)
plt.show()
```
